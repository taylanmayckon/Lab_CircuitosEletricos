```python
import matplotlib.pyplot as plt
import numpy as np

def plot_circuito(RL, PL, IL, RTH):
    fig, ax1 = plt.subplots()

    RL = np.array(RL)
    PL = np.array(PL)
    IL = np.array(IL)

    # Índice do ponto mais próximo de RTH
    idx_rth = np.argmin(np.abs(RL - RTH))
    RL_rth = RL[idx_rth]
    PL_rth = PL[idx_rth]
    IL_rth = IL[idx_rth]

    # Eixo primário (Corrente)
    ax1.set_xlabel('Resistência de carga (Ω)')
    ax1.set_ylabel('Corrente (mA)', color='black')
    ax1.plot(RL, IL * 1000, color='gray', linestyle='-', label='Corrente IL')
    ax1.tick_params(axis='y', labelcolor='black')

    # Eixo secundário (Potência)
    ax2 = ax1.twinx()
    ax2.set_ylabel('Potência (mW)', color='black')
    ax2.plot(RL, PL * 1000, color='tab:red', linestyle='--', label='Potência PL')
    ax2.tick_params(axis='y', labelcolor='black')

    # Ponto de Pmax sobre a linha da potência
    ax2.plot(RL_rth, PL_rth * 1000, 'o', color='blue', label='Ponto de Pmax')

    # Anotação abaixo do ponto de máxima potência
    ax2.annotate(f'Pmax ≈ {PL_rth*1000:.2f} mW\nRL ≈ {RL_rth} Ω\nIL ≈ {IL_rth*1000:.2f} mA',
                 xy=(RL_rth, PL_rth * 1000),
                 xytext=(0, -12),
                 textcoords='offset points',
                 ha='center',
                 va='top',
                 fontsize=9,
                 color='blue',
                 bbox=dict(boxstyle="round,pad=0.2", fc="white", ec="blue", lw=0.5))

    # Combina as legendas dos dois eixos
    lines1, labels1 = ax1.get_legend_handles_labels()
    lines2, labels2 = ax2.get_legend_handles_labels()
    ax1.legend(lines1 + lines2, labels1 + labels2, loc='best')

    plt.title('Corrente e Potência vs Resistência de carga')
    plt.grid(True)
    plt.tight_layout()
    plt.show()
```

# Questão 1 - Circuito 1

```python
VTH_1_1 = 2.266 # VTH do circ 1
RTH_1_1 = 939.08 # RTH do circ 1
IL_1_1 = list(map(lambda RL: VTH_1_1/(RTH_1_1 + RL), range(600, 1200)))  # Dados de corrente (offset de 300 +/-)
RL_1_1 = [i for i in range(600, 1200)] # Resistência da carga calculada
PL_1_1 = [IL_1_1[i] * IL_1_1[i] * RL_1_1[i] for i in range(len(RL_1_1))] # Potências

plot_circuito(RL_1_1, PL_1_1, IL_1_1, RTH_1_1);

```
